```python
df4["Age_band"] = np.where(df4['Age']>=25, 'Old', 'Young') 
#If Age is same and more than 25, i replace old, and less than 25, i replace young in Age_band column.
```

```python
df4['TypingTime_band'] = pd.qcut(df4.TypingTime, q=3, labels = ['Slow','Midium','Fast'])
#TypingTime is divided in Slow, Midium, Fast
```
